```python
# Rayyan Dirie - Dedrone Data Science Internship - Technical Assessment
#Data Visualization - Regional Protest Statistics
import pandas as pd
import matplotlib.pyplot as plt

#Create Dataframe for Protest Data
protest_data = pd.read_csv("/Users/rayyandirie/Documents/Dedrone_Docs/North_America-United_States.csv")
#Define Regions
new_england = ["Connecticut", "Maine", "Massachusetts", "New Hampshire", "Rhode Island", "Vermont"]
middle_atlantic = ["New Jersey", "New York", "Pennsylvania"]
east_north_central = ["Illinois", "Indiana", "Michigan", "Ohio", "Wisconsin"]
west_north_central = ["Iowa", "Kansas", "Minnesota", "Missouri", "Nebraska", "North Dakota", "South Dakota"]
south_atlantic = ["Delaware", "Florida", "Georgia", "Maryland", "North Carolina", "South Carolina", "Virginia", "District of Columbia", "West Virginia"]
east_south_central = ["Alabama", "Kentucky", "Mississippi", "Tennessee"]
west_south_central = ["Arkansas", "Louisiana", "Oklahoma", "Texas"]
mountain = ["Arizona", "Colorado", "Idaho", "Montana", "Nevada", "New Mexico", "Utah", "Wyoming"]
pacific = ["Alaska", "California", "Hawaii", "Oregon", "Washington"]

regions = [new_england, middle_atlantic, east_north_central, west_north_central, south_atlantic, 
           east_south_central, west_south_central, mountain, pacific]
region_names = ["new_england", "middle_atlantic", "east_north_central", "west_north_central", "south_atlantic", 
           "east_south_central", "west_south_central", "mountain", "pacific"]

#Define Region Totals
new_england_count = 0
middle_atlantic_count = 0
east_north_central_count = 0
west_north_central_count = 0
south_atlantic_count = 0
east_south_central_count = 0
west_south_central_count = 0
mountain_count = 0
pacific_count = 0

#Gather Data for Individual State Pie Chart
protests_per_state = protest_data['admin1'].value_counts()

#Calculate protests per region using an parralel iterative loop (accessing regions and their respective region name strings)
for n in protests_per_state.index:
    for regionList, regionName in zip(regions, region_names):
        if n in regionList:
            exec(f"{regionName}_count += {protests_per_state[n]}")
            
region_counts = [new_england_count, middle_atlantic_count, east_north_central_count, west_north_central_count, south_atlantic_count,
                 east_south_central_count, west_south_central_count, mountain_count, pacific_count]

#Calculate Average Number of Protests per State for each Region
region_average = []
for group, count in zip(regions, region_counts):
    region_average.append('{:.2f}'.format((count / len(group))))
# #Print and Chart Data
print("Census Region Data")
for name, count, avg in zip(region_names, region_counts, region_average):
    print(f"{name}: {count} - Average Number of Protests per State: {avg}")
plt.figure(figsize=(6, 6))
plt.pie(region_counts, labels=region_names, autopct='%1.1f%%')
plt.title('Distribution of Protests Across Census State Regions\n')
plt.axis('equal') 
plt.show()
```
